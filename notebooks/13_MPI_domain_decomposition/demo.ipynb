```python
np=4
! mpirun -np $np python3 hello_world_mpi.py

from ipywidgets import HBox, Output
children = []
for i in range(np):
    output = Output()
    with output:
        display(SVG(filename=f"out.{i}.svg"))
    children.append(output)
display(HBox(children))
```
